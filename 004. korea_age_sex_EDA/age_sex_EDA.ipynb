# 확진자수 막대그래프 자료 

```
import pandas as pd
data = {'male': [58,242,1103,388,353,501,423,243,119],
        'female':[43,218,1293,521,868,1190,709,352,273]}
age = pd.DataFrame(data)
age = age.T
col=['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-']
age.columns=col 
age = age.T
age
```

```
import matplotlib.pyplot as plt
plt.figure()
plt.barh(age.index, age.male, color='blue', label='male')
plt.barh(age.index, -age.female, color='red', label='female')
plt.xlabel('patient')
plt.ylabel('age')
plt.legend()
plt.show()
```

```
age = pd.read_csv('/content/daily_Confirmed.csv') #구글 코랩실행path
age.index=age['Date']
col2=['0-9','10-','20-29','30-39','40-49','50-59','60-69','70-79','80-']
age = age[col2]
age.columns = [col]
age = age.iloc[4:-1, :]
age
```

```
latest.iloc[0,0]
```

```
age_sum = latest['confirmed'].sum()
rate=[]
for i in range(len(col)):
  a = latest.iloc[i,0] /age_sum *100
  rate.append(a)

latest['rate'] = rate
latest
```

```
plt.rcParams['figure.figsize']=8,4
color = 'tab:red'
ax= latest['confirmed'].plot( title = " nCOV19 - Confirmed individule by Age", rot=0,color=color)
ax.set_ylabel('Confirmed_count',color=color)
ax.tick_params(axis='y',color=color)
ax.legend()

ax2 = ax.twinx()
color = 'tab:blue'
ax2 = latest['rate'].plot(kind='bar', color=color, alpha=0.1)
ax2.set_ylabel(' %')

for p in ax2.patches: 
  left1, bottom1, width1, height1 = p.get_bbox().bounds 
  ax2.annotate("%.f%%"%(height1), xy=(left1+width1/2, bottom1+height1/2), ha='center', va='center')

plt.sca(ax)
plt.show()
```

#  확진자수 데이터셋에서 성별 분포만 뽑아서 시각화

```
age = pd.read_csv('/content/daily_Confirmed.csv')
age.index = age['Date']
age_by_sex = age[['Female', 'Male','Sum']]
age_by_sex = age_by_sex.iloc[:-1,:]
age_by_sex
```

```
latest = age_by_sex.iloc[-1, :-1]
latest = pd.DataFrame(latest)
latest.columns=['confirmed']
latest
```

```
age_sum = latest['confirmed'].sum()
index =  ['Female',',Male']
rate=[]
for i in range(len(index)):
  a = latest.iloc[i,0] / age_sum *100
  rate.append(a)

latest['rate'] = rate
latest
```

```
plt.rcParams['figure.figsize']=12,8
group_explodes = (0.05, 0)
plt.pie(latest['rate'], explode=group_explodes,labels=["female","male"], colors=['lightcoral', 'lightskyblue'],autopct='%1.2f%%', shadow=True, startangle=90, textprops={'fontsize':14})
plt.axis('equal')
plt.title('Confirmed individual by sex')
plt.show()
```
